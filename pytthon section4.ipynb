```python
#programme=51
sum=0
r=[]
q=[]
print("house floor area calculator.")
print("please enter the number of rooms")
a=int(input())
for i in range(a):
    s=input("please enter the name of the room"+str(i+1))
    r.append(s)
    z=float(input("please enter the length of the room"+str(i+1)))
    d=float(input("please enter the width of the room"+str(i+1)))
    area=z*d
    e=round(area,1)
    sum=sum+e
    q.append(e)
print("house details")
for loop in range(len(r)):
    print("room-",r[loop])
    print(q[loop],"meters squared")
print("total area is calculated as:")
print(sum,"meters squared")
    
```

```python
#programme=53
q=["empty"]*5
w=["empty"]*5
e=["empty"]*5
r=["empty"]*5
print("how many members have requested a seat ? ")
a=int(input())
for loop in range(a):
    print("please enter your name")
    b=input()
    print("which row would you like to sit in ?")
    c=int(input())
    print("which seat number would you like to sit in?")
    d=int(input())
    print("Thank you ,your name has been added")
    if d==1:
        q.pop(c-1)
        q.insert(c-1,b)
    elif d==2:
        w.pop(c-1)
        w.insert(c-1,b)
    elif d==3:
        e.pop(c-1)
        e.insert(c-1,b)
    else:
        r.pop(c-1)
        r.insert(c-1,b)
print("bus seats have been booked as follows") 
for i in range(5):
    
    print(str(i),q[i],w[i],e[i],r[i])
```
